```python
import warnings; warnings.filterwarnings('ignore')
%matplotlib inline
mydpi=120

from sklearn.datasets import load_boston
from sklearn.datasets import load_diabetes
import numpy as np

import sys
sys.path.append('..')
from korr import pearson, bracket_pval
```

## Load demo dataset

```python
x = load_boston().data[:,[0,1,2,4,5,6,7,8,9,10,11,12]]
#x = load_diabetes().data
r, pval = pearson(x)
```

## Default Settings

```python
bracket_pval(r, pval, dpi=mydpi);
```

## Manual p-value brackets

```python
bracket_pval(r, pval, plim=[.00001], dpi=mydpi);
```

```python
bracket_pval(r, pval, plim=[.000050, .0000025, .0000001], dpi=mydpi);
```

## Specific number of brackets of 10-multiples


```python
bracket_pval(r, pval, plim=4, dpi=mydpi);
```

```python
bracket_pval(r, pval, plim=6, dpi=mydpi);
```

```python
bracket_pval(r, pval, plim=8, dpi=mydpi);
```

## Number of Bins

```python
bracket_pval(r, pval, plim=6, nbins=7, dpi=mydpi);
```

## Matrix or Vector?
Both are eligible. 
However a correlation matrix `r` contains each correlation twice and the diagonals. 
The function `bracket_pval` would only use the triangular matrix without diagonals by default.

```python
idx = (np.tri(N=r.shape[0], k=-1) == 1);
bracket_pval(r[idx], pval[idx], plim=6, nbins=7, dpi=mydpi);
```
