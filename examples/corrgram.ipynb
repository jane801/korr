```python
#import numpy as np
from sklearn.datasets import load_diabetes

import sys
sys.path.append('..')
from korr import pearson, corrgram
```

Load demo dataset

```python
data = load_diabetes()
X = data.data
varnames = data.feature_names
```

Compute correlations

```python
rho, pval = pearson(X)
```

The correlation coefficients are plotted with + and - prefix. The corresponding p-values are below in brackets.

```python
corrgram(rho, pval);
```

Use `dpi` to change the resolution. 
Pick another [color theme](https://matplotlib.org/examples/color/colormaps_reference.html).

```python
corrgram(rho, pval, dpi=60, cmap='PuOr');
```

Add the variable names to the plot axis

```python
corrgram(rho, pval, varnames, dpi=120);
```
